```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# przygotowanie danych
from sklearn.preprocessing import StandardScaler, MinMaxScaler
dataset = pd.read_csv(r'titanic.csv')
dataset.head()
```

```python
value = dataset.loc[dataset['Fare'].notnull(), 'Fare'].values
value.shape
```

```python
plt.figure(figsize=(20,10))
plt.hist(value, bins=100)
plt.show()
```

```python
#średnia

average = np.round(np.mean(value))
average
```

```python
#mediana
median = np.median(value)
median
```

```python
#moda

mode = stats.mode(value)
mode
```

```python
#kwartyle

quartile_0 = np.quantile(value, 0.0)
quartile_1 = np.round(np.quantile(value, 0.25))
quartile_3 = np.round(np.quantile(value, 0.75))
quartile_4 = np.round(np.quantile(value, 1.0))
print(f'Q0: {quartile_0}')
print(f'Q1: {quartile_1}')
print(f'Q2: {median}')
print(f'Q3: {quartile_3}')
print(f'Q4: {quartile_4}')
```

```python
plt.figure(figsize=(20,10))
plt.hist(value, bins=100)
plt.axvline(x = average, color='red', label = 'Średnia')
plt.axvline(x = median, color='green', label = 'Mediana')
plt.axvline(x = mode[0], color='yellow', label = 'Moda')
plt.axvline(x = quartile_0, color='black', label = 'Min',alpha=0.5)
plt.axvline(x = quartile_1, color='black', label = 'Q1',alpha=0.5)
plt.axvline(x = quartile_3, color='black', label = 'Q3',alpha=0.5)
plt.axvline(x = quartile_4, color='black', label = 'Max',alpha=0.5)
plt.legend(loc='upper right')
plt.show()
```
